```python
import pandas as pd 
import RNA

df = pd.read_excel("../datasets/gRNA list for MFE2.xlsx", skiprows=2)
df.tail()
```

```python
mer_motif1 = [RNA.fold(seq.upper().replace("T", "U")) if type(seq)==str else None for seq in df["motif1"]]
mer_motif2 = [RNA.fold(seq.upper().replace("T", "U")) if type(seq)==str else None for seq in df["motif2"]]
mer_motif2
```

```python
df["motif1_mfe_structure"] = [l[0] if type(l) == list else None for l in mer_motif1]
df["motif1_mfe"] = [l[1] if type(l) == list else None for l in mer_motif1]

df["motif2_mfe_structure"] = [l[0] if type(l) == list else None for l in mer_motif2]
df["motif2_mfe"] = [l[1] if type(l) == list else None for l in mer_motif2]
```

```python
df.to_excel("../outputs/gRNA list for MFE2_RNAfold.xlsx")
```
